```python
import numpy as np
import matplotlib.pyplot as plt


# 设置中文字体为黑体（SimHei），解决中文乱码
plt.rcParams['font.sans-serif'] = ['SimHei']  
# 解决负号 '-' 显示为方块的问题
plt.rcParams['axes.unicode_minus'] = False

# 模拟数据
np.random.seed(42)
X = np.linspace(0, 10, 10)
y = 2.5 * X + 5 + np.random.normal(0, 3, size=len(X))

# OLS 拟合
coef = np.polyfit(X, y, 1)  # [斜率, 截距]
y_pred = np.polyval(coef, X)

# 绘图
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='观测数据')
plt.plot(X, y_pred, color='red', label=f'OLS拟合直线: y={coef[0]:.2f}x+{coef[1]:.2f}')

# 画残差
for i in range(len(X)):
    plt.vlines(X[i], y_pred[i], y[i], color='green', linestyle='--', alpha=0.7)

plt.xlabel('自变量 X')
plt.ylabel('因变量 Y')
plt.title('OLS（普通最小二乘法）原理示意图')
plt.legend()
plt.grid(True)
plt.show()
```
